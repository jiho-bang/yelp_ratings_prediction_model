```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import confusion_matrix
```

# Problem 2: Predicting Yelp Ratings

```python
train = pd.read_csv('yelp142_train.csv')
test = pd.read_csv('yelp142_test.csv')

train.head()
```

## 2a)

This modeling approach is reasonable. The missing data can be interpreted as a special variable where no information is available. In other words, the "(Missing)" categorical variable would just tell you whether there was information available for this variable or not. Moreover, this approach provides some flexibility by potentially conserving some partial information that is available rather than outright discarding all the data. Finally, the fact of not having information itself can be a potential indicator of the quality of the restaurant.

## 2b) i)

Below is the code for building my linear regression model predicting stars based on all the features, using "(Missing)" as the reference label. 

```python
ols = smf.ols(formula = "stars ~ review_count + \
                        C(GoodForKids, Treatment(reference = '(Missing)')) + \
                        C(Alcohol, Treatment(reference = '(Missing)')) + \
                        C(BusinessAcceptsCreditCards, Treatment(reference = '(Missing)')) + \
                        C(WiFi, Treatment(reference = '(Missing)')) + \
                        C(BikeParking, Treatment(reference = '(Missing)')) + \
                        C(ByAppointmentOnly, Treatment(reference = '(Missing)')) + \
                        C(WheelechairAccessible, Treatment(reference = '(Missing)')) + \
                        C(OutdoorSeating, Treatment(reference = '(Missing)')) + \
                        C(RestaurantsReservations, Treatment(reference = '(Missing)')) + \
                        C(DogsAllowed, Treatment(reference = '(Missing)')) + \
                        C(Caters, Treatment(reference = '(Missing)'))", data = train)
model = ols.fit()
print(model.summary())
```

## 2b) ii)

```python
y_train = train['stars']
x_train = pd.get_dummies(train.drop(['stars'], axis = 1))
y_test = test['stars']
x_test = pd.get_dummies(test.drop(['stars'], axis = 1))

vals = {'ccp_alpha': np.linspace(0, 0.10, 201)}
dtr = DecisionTreeRegressor()
cv = KFold(n_splits = 5, random_state = 1, shuffle = True)
dtr_cv = GridSearchCV(dtr, param_grid = vals, scoring = 'r2', cv = cv, verbose = 0)
dtr_cv.fit(x_train, y_train)
```

```python
acc = dtr_cv.cv_results_['mean_test_score'] 
ccp = dtr_cv.cv_results_['param_ccp_alpha'].data

pd.DataFrame({'ccp alpha' : ccp, 'Validation Accuracy': acc}).head(20)
```

```python
print('The best ccp_alpha is', dtr_cv.best_params_)
print('The best score (accuracy): ', dtr_cv.best_score_)
```

Above is the code for building my regression tree model. To do the cross validation, I first specified a grid with different ccp_alpha values. I then initialized a DecisionTreeRegressor and used the KFold function to split my data randomly in k-folds. Moreover, I made sure to score my GridSearchCV using 'r2.' Finally, I chose the best value of 'ccp_alpha' by looking at the best_params attribute which outputs the highest validation scoring. As you can see, it matches the output of the table.

## 2b) iii)

```python
def OSR2(model, X_test, y_test, y_train):
    
    y_pred = model.predict(X_test)
    SSE = np.sum((y_test - y_pred)**2)
    SST = np.sum((y_test - np.mean(y_train))**2)
                 
    return (1 - SSE/SST)

def OSR2_lin(model, df_train, df_test, dependent_var):   
    y_test = df_test[dependent_var]
    y_pred = model.predict(df_test)
    SSE = np.sum((y_test - y_pred)**2)
    SST = np.sum((y_test - np.mean(df_train[dependent_var]))**2)    
    return 1 - SSE/SST

y_pred = model.predict(test)
y_treepred = dtr_cv.predict(x_test)
y_true = test['stars']
```

```python
print('Linear Regression Model OSR2', OSR2_lin(model, train, test, 'stars'))
print('Regression Tree Model OSR2', OSR2(dtr_cv, x_test, y_test, y_train))
```

```python
print('Linear Regression Model MAE', mean_absolute_error(y_true, y_pred))
print('Regression Tree Model MAE', mean_absolute_error(y_test, y_treepred))
```

Above we can see the values for OSR2 and MAE for both the tree and linear regression models. A negative OSR2 is usually an indicator that doesn't work well on the test set. Although the OSR2s are both positive, given that our OSR2s are both pretty low and our MAEs are relatively higher, we can conclude that the performance of the two models is not too good.

## 2c)

Below is the code for creating the fourOrAbove variable in both the training and test datasets.

```python
four_train = [1 if i >= 4.0 else 0 for i in train['stars']]
four_test = [1 if i >= 4.0 else 0 for i in test['stars']]
train4 = train.copy(deep = True)
test4 = test.copy(deep = True)
train4['fourOrAbove'] = four_train
test4['fourOrAbove'] = four_test
train4.drop(['stars'], axis = 1, inplace = True)
test4.drop(['stars'], axis = 1, inplace = True)

train4.head()
```

## 2d) i)

This modeling choice is reasonable. It is unreasonable to weight false positives and false negatives equally when we know that one is significantly costlier than the other. However, in this instance, there is no indication that false positives and false negatives are weighted entirely different. This tells us that none is more costly than the other, so weighing them equally should not be problem. 

## 2d) ii)

Below is the code for converting the predictions of the regression models to predictions of fourOrAbove.

```python
y_pred_ols = model.predict(test.drop(['stars'], axis = 1))
lin_threshold = [1 if i >= 4.0 else 0 for i in y_pred_ols]

y_pred_tree = dtr_cv.predict(x_test)
tree_threshold = [1 if i >= 4.0 else 0 for i in y_pred_tree]

print(str(lin_threshold[:100]))
print('\n')
print(str(tree_threshold[:100]))
```

## 2d) iii)

Below is the code for building my logistic regression model based on all the features, using "(Missing)" as the reference label. 

```python
logistic = smf.logit(formula = "fourOrAbove ~ review_count + \
                        C(GoodForKids, Treatment(reference = '(Missing)')) + \
                        C(Alcohol, Treatment(reference = '(Missing)')) + \
                        C(BusinessAcceptsCreditCards, Treatment(reference = '(Missing)')) + \
                        C(WiFi, Treatment(reference = '(Missing)')) + \
                        C(BikeParking, Treatment(reference = '(Missing)')) + \
                        C(ByAppointmentOnly, Treatment(reference = '(Missing)')) + \
                        C(WheelechairAccessible, Treatment(reference = '(Missing)')) + \
                        C(OutdoorSeating, Treatment(reference = '(Missing)')) + \
                        C(RestaurantsReservations, Treatment(reference = '(Missing)')) + \
                        C(DogsAllowed, Treatment(reference = '(Missing)')) + \
                        C(Caters, Treatment(reference = '(Missing)'))", data = train4)
logistic = logistic.fit()
print(logistic.summary())
```

## 2d) iv)

```python
y_train4 = train4['fourOrAbove']
x_train4 = pd.get_dummies(train4.drop(['fourOrAbove'], axis = 1))
y_test4 = test4['fourOrAbove']
x_test4 = pd.get_dummies(test4.drop(['fourOrAbove'], axis = 1))

dtc = DecisionTreeClassifier()
cv = KFold(n_splits = 5, random_state = 1, shuffle = True)
dtc_cv_acc = GridSearchCV(dtc, param_grid = vals, scoring = 'accuracy', cv = cv, verbose = 0)
dtc_cv_acc.fit(x_train4, y_train4)
```

```python
acc2 = dtc_cv_acc.cv_results_['mean_test_score'] 
ccp2 = dtc_cv_acc.cv_results_['param_ccp_alpha'].data

pd.DataFrame({'ccp alpha' : ccp2, 'Validation Accuracy': acc2}).head(20)
```

```python
print('The best ccp_alpha is', dtc_cv_acc.best_params_)
print('The best score (accuracy): ', dtc_cv_acc.best_score_)
```

```python
print('Node count =', dtc_cv_acc.best_estimator_.tree_.node_count)

plt.figure(figsize=(20,10))
plot_tree(dtc_cv_acc.best_estimator_, 
          feature_names=x_train4.columns.values.tolist(), 
          class_names=['0','1'],
          filled=True,
          impurity=True,
          rounded=True,
          fontsize=12,
          max_depth=3)
plt.show()
```

Above is the code for building my classification tree model. To do the cross validation, I followed a similar procedure to what I did in $2b)ii)$. I first specified a grid with different ccp_alpha values. I then initialized a DecisionTreeClassifier and used the KFold function to split my data randomly in k-folds. Moreover, I made sure to score my GridSearchCV using 'accuracy.' Finally, I chose the best value of 'ccp_alpha' by looking at the best_params attribute which outputs the highest scoring. 

## 2d) v)

Below is the code for building a table that evaluates the performance of all the models I've built.

```python
def performance(model, x_test, y_test):
    y_pred = model.predict(x_test)
    cm = confusion_matrix(y_test, y_pred)
    acc = (cm.ravel()[0] + cm.ravel()[3]) / sum(cm.ravel())
    TPR = cm.ravel()[3]/(cm.ravel()[3] + cm.ravel()[2])
    FPR = cm.ravel()[1]/(cm.ravel()[1] + cm.ravel()[0])
    return [acc, TPR, FPR]
    

cm_baseline = confusion_matrix(test4['fourOrAbove'], [0] * test.shape[0])
bacc = (cm_baseline.ravel()[0] + cm_baseline.ravel()[3]) / sum(cm_baseline.ravel())
bTPR = cm_baseline.ravel()[3]/(cm_baseline.ravel()[3] + cm_baseline.ravel()[2])
bFPR = cm_baseline.ravel()[1]/(cm_baseline.ravel()[1] + cm_baseline.ravel()[0])

lincm = confusion_matrix(y_test4, lin_threshold)
linacc = (lincm.ravel()[0] + lincm.ravel()[3]) / sum(lincm.ravel())
linTPR = lincm.ravel()[3]/(lincm.ravel()[3] + lincm.ravel()[2])
linFPR = lincm.ravel()[1]/(lincm.ravel()[1] + lincm.ravel()[0])

tcm = confusion_matrix(y_test4, tree_threshold)
tacc = (tcm.ravel()[0] + tcm.ravel()[3]) / sum(tcm.ravel())
tTPR = tcm.ravel()[3]/(tcm.ravel()[3] + tcm.ravel()[2])
tFPR = tcm.ravel()[1]/(tcm.ravel()[1] + tcm.ravel()[0])

log_pred = logistic.predict(test)
log_threshold = [1 if i >= 0.5 else 0 for i in log_pred]
logcm = confusion_matrix(y_test4, log_threshold)
logacc = (logcm.ravel()[0] + logcm.ravel()[3]) / sum(logcm.ravel())
logTPR = logcm.ravel()[3]/(logcm.ravel()[3] + logcm.ravel()[2])
logFPR = logcm.ravel()[1]/(logcm.ravel()[1] + logcm.ravel()[0])
```

```python
table = pd.DataFrame(
    columns=[' ', 'Baseline', 'Linear - Threshold', 'Tree - Threshold', 'Logistic', \
             'Classification Tree' ])
table[' '] = ['Accuracy', 'TPR', 'FPR']
table['Classification Tree'] = performance(dtc_cv_acc, x_test4, y_test4)
table['Baseline'] = [bacc, bTPR, bFPR]
table['Linear - Threshold'] = [linacc, linTPR, linFPR]
table['Tree - Threshold'] = [tacc, tTPR, tFPR]
table['Logistic'] = [logacc, logTPR, logFPR]

table
```

The results outputted by the table seem reasonable. All the models have relatively similar values, each with their pros and cons. The performance is all pretty decent. All the accuracies outdo the baseline accuracy by a good amount. 

There are also some tradeoff between the models. The baseline model, for example, has a 0 for both TPR and FPR but that is because we know we are purposefully assigning everything to 0. Moreover, we see that the higher the accuracy, the more suceptible it is to having higher FPR. This makes sense since the tradeoff for assigning more trues is that you're also more likely to increase the outcomes of FPR. 

Finally, I would recommend the classification tree since it seems to have the best performance. It has close to the highest accuracy and TPR while maintaining a pretty low FPR.
